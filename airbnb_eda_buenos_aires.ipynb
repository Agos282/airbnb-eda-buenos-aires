Se decidió aplicar un filtro de precios menores a 10000 para mejorar la calidad del análisis

## Conclusiones

A lo largo del análisis exploratorio de datos (EDA) sobre los alojamientos de Airbnb en Buenos Aires, se identificaron varios aspectos relevantes:

- La distribución de precios presenta una fuerte asimetría debido a la presencia de valores atípicos (outliers), los cuales fueron filtrados para obtener un análisis más representativo.

- Los alojamientos completos (Entire home/apartment) son los más comunes y presentan precios promedio significativamente más altos en comparación con habitaciones privadas o compartidas.

- Barrios como Palermo y Recoleta concentran la mayor cantidad de alojamientos, lo que sugiere una alta demanda y competencia en estas zonas.

- Puerto Madero se destaca como el barrio con precios más elevados, aunque algunos valores pueden estar influenciados por propiedades de lujo o datos atípicos.

- No se observa una relación directa entre la cantidad de reviews y el precio, lo que indica que los alojamientos más caros no necesariamente son los más populares.

En conclusión, el mercado de Airbnb en la ciudad muestra una alta concentración en zonas turísticas, con una amplia variabilidad de precios según el tipo de alojamiento y la ubicación.

## 📅 Análisis de disponibilidad y precios en el tiempo 

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df_calendar = pd.read_csv("calendar.csv") 
```

## Precio

```python
df_calendar["price"] = df_calendar["price"].replace("[$,]", "", regex=True).astype(float) 
```

## Fecha

```python
df_calendar["date"] = pd.to_datetime(df_calendar["date"])
```

## Disponibilidad

```python
df_calendar["available"] = df_calendar["available"].map({"t": 1, "f": 0})
```

## 📊 1. PRECIO EN EL TIEMPO

```python
price_by_date = df_calendar.groupby("date")["price"].mean()

price_by_date.plot(figsize=(10,5))
plt.title("Precio promedio por fecha") 
plt.xlabel("Fecha") 
plt.ylabel("Precio") 
plt.show()
```

## 📊 2. DISPONIBILIDAD

```python
availability_by_date = df_calendar.groupby("date")["available"].mean() 

availability_by_date.plot(figsize=(10,5)) 
plt.title("Disponibilidad promedio en el tiempo") 
plt.xlabel("Fecha") 
plt.ylabel("Disponibilidad") 
plt.show()
```

## 💡 3. CRUCE CLAVE

```python
df_calendar.groupby("available")["price"].mean() 
```

## 📊 Gráfico por mes

```python
df_calendar["month"] = df_calendar["date"].dt.month

df_calendar.groupby("month")["price"].mean().plot()
plt.title("Precio promedio por mes")
plt.xlabel("Mes")
plt.ylabel("Precio")
plt.show()
```

### Insight
👉Precio en el tiempo

Se observa un aumento en los precios hacia fin de año e inicios del siguiente, lo que podría estar relacionado con temporadas de alta demanda como vacaciones o eventos.    

👉Disponibilidad

La disponibilidad disminuye a lo largo del tiempo, lo que indica un aumento en la ocupación de los alojamientos.

👉Cruce clave

Los alojamientos no disponibles (ocupados) presentan un precio promedio más alto que aquellos disponibles, lo que sugiere que las propiedades más costosas tienden a tener mayor demanda.


## Análisis de reviews y comportamiento de usuarios

```python
df_reviews = pd.read_csv("reviews.csv")
```

```python
df_reviews.head()
df_reviews.info() 
```

## LIMPIEZA
👉 Fecha

```python
df_reviews["date"] = pd.to_datetime(df_reviews["date"])
```

## 👉 Comentarios

```python
df_reviews["comments"] = df_reviews["comments"].astype(str) 
```

## 📊 ACTIVIDAD EN EL TIEMPO

👉 Cantidad de reviews por fecha:

```python
reviews_by_date = df_reviews.groupby("date").size() 

reviews_by_date.plot(figsize=(10,5)) 
plt.title("Cantidad de reviews en el tiempo") 
plt.xlabel("Fecha") 
plt.ylabel("Cantidad de reviews") 
plt.show()
```

## Insight

Se observa un aumento en la cantidad de reviews en determinados períodos, lo que podría indicar mayor actividad o demanda.

## Unir con Listings

```python
df_merged =df_reviews.merge(df1, left_on="listing_id", right_on="id") 
```

## 📊 Reviews por tipo de alojamiento

```python
df_merged.groupby("room_type").size().plot(kind="bar") 

plt.title("Cantidad de reviews por tipo de alojamiento")
plt.xlabel("Tipo de alojamiento") 
plt.ylabel("Cantidad de reviews")
plt.show() 
```

## 📊 Reviews vs precio

```python
df_merged.plot(kind="scatter", x="number_of_reviews", y="price") 

plt.title("Relacion entre reviews y precio") 
plt.show() 
```

### Insight

Se observa que los alojamientos con mayor cantidad de reviews no necesariamente son los más caros, lo que sugiere que la popularidad no depende únicamente del precio.

Además, ciertos tipos de alojamiento concentran mayor cantidad de reseñas, indicando una preferencia por parte de los usuarios.
